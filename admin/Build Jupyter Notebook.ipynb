```python
import nbformat
import shutil
import os
import glob

def add_header_footer(input_nb_path, header_nb_path, footer_nb_path, output_nb_path):
    # Load the main notebook
    with open(input_nb_path, 'r') as nb_file:
        nb = nbformat.read(nb_file, nbformat.NO_CONVERT)
    
    # Load the header notebook
    with open(header_nb_path, 'r') as header_file:
        header_nb = nbformat.read(header_file, nbformat.NO_CONVERT)
    
    # Load the footer notebook
    with open(footer_nb_path, 'r') as footer_file:
        footer_nb = nbformat.read(footer_file, nbformat.NO_CONVERT)
    
    # Insert the header cells at the beginning of the main notebook
    nb.cells = header_nb.cells + nb.cells
    
    # Append the footer cells at the end of the main notebook
    nb.cells += footer_nb.cells
    
    # Clear all output in the notebook
    for cell in nb.cells:
        cell.outputs = []
    
    # Hide the cell with name "Answer"
    for cell in nb.cells:
        # if cell.cell_type == 'markdown' and "Answer" in cell.source:
        #         cell.metadata['hide_input'] = True
        if cell.cell_type == 'markdown' and "Answer" in cell.source:
            cell.metadata['jp-MarkdownHeadingCollapsed'] = True

        if cell.cell_type == 'markdown' and "../images" in cell.source:
            cell.source = cell.source.replace("../images", "./images")
            
    # Save the modified notebook with the output file path
    with open(output_nb_path, 'w') as nb_file:
        nbformat.write(nb, nb_file)

# Example usage
header_nb_path = './brand_header.ipynb'
footer_nb_path = './brand_footer.ipynb'


# Directory containing the .ipynb files
input_directory = '/home/pi/raspberry-pi-beginner/original'
output_directory = '/home/pi/raspberry-pi-beginner'

# Get all .ipynb file paths in the directory
ipynb_files = glob.glob(os.path.join(input_directory, '*.ipynb'))

# Loop through the .ipynb files
for input_nb_path in ipynb_files:
    # Extract the filename with the extension
    file_name = os.path.basename(input_nb_path)

    output_nb_path = f"{output_directory}/{file_name}"
    
    # Process the .ipynb file
    print("Processing file:", input_nb_path)
    print("Filename:", file_name)
    print("Output:", output_nb_path)
    
    add_header_footer(input_nb_path, header_nb_path, footer_nb_path, output_nb_path)
```
